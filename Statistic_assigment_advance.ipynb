1.What is a random variable in probability theory?

In probability theory, a random variable is a variable that takes on different numerical values based on the outcome of a random experiment.

Formally, a random variable is a function that assigns a real number to each possible outcome in a sample space.

🔹 Two Types of Random Variables

Discrete Random Variable

Takes on countable or finite values.

Example:

Number of heads when flipping a coin three times → {0, 1, 2, 3}

Number of customers arriving in a store per hour.

Continuous Random Variable

Takes on infinitely many possible values within an interval.

Example:

The exact height of a person (e.g., 170.5 cm)

The time (in seconds) it takes for a computer to process a task.

2. What are the types of random variables)

In probability theory, there are two main types of random variables:

🔹 1. Discrete Random Variable

A discrete random variable can take only a finite or countably infinite number of possible values.

Key Features:

Values are distinct and separate (you can list them).

Probabilities are assigned to each possible value using a probability mass function (PMF).

The sum of all probabilities equals 1.

 Examples:

Number of heads when flipping 3 coins → {0, 1, 2, 3}

Number of students present in a class

Number of phone calls received in an hour

🔹 2. Continuous Random Variable

A continuous random variable can take any value within a given range (uncountably infinite).

 Key Features:

Values are not countable (infinitely many within an interval).

Described using a probability density function (PDF) instead of a PMF.

The probability of taking an exact value is zero; probabilities are found over intervals.

Here are the key properties of a Bernoulli distribution:

•	Probability Mass Function (PMF): The probability of observing an outcome $x$ is given by:

o	If 14$x=1$ (success), 15$P(X=1) = p^1 (1-p)^0 = p$.16
o	If 17$x=0$ (failure), 18$P(X=0) = p^0 (1-p)^1 = 1-p = q$.

Key Properties and FormulaTo use the binomial distribution, an experiment must meet the following four criteria (often called the BINS conditions):Condition                    Description

            The outcome of one trial must not affect the outcome of any other trial.
Number of Trials (Fixed)
                The total number of trials, $n$, must be fixed in advance.
                
Same Probability
             The probability of success, $p$, must remain constant for every trial.

13.What is the Central Limit Theorem (CLT), and why is it critical in statistics?

The Central Limit Theorem (CLT) is a cornerstone of modern statistics. It states that if you take a sufficiently large number of independent random samples from any population (regardless of its original distribution—be it uniform, skewed, binomial, etc.), the distribution of the sample means will tend to follow a Normal Distribution (bell curve)

Why the CLT is Critical in Statistics

The Central Limit Theorem is critical because it allows statisticians and analysts to use the powerful tools of the Normal Distribution to make inferences about a population, even when the original population's distribution is unknown or non-normal.

Foundation for Inferential Statistics:

Confidence Intervals: It allows us to construct confidence intervals around a sample mean, which provides a range of values likely to contain the true population mean (21$\mu$).

Hypothesis Testing: It is the theoretical basis for most parametric hypothesis tests (like 23$t$-tests and Z-tests), enabling us to test claims about population means.

Simplification of Real-World Data: In real-world data collection, the distribution of individual data points is often complex or unknown. The CLT provides a reliable way to simplify this chaos by stating that the average behavior of repeated samples will follow a predictable normal pattern.

Applicability: It makes statistical inference applicable to a vast number of fields, including polling, quality control, finance, and scientific research, by ensuring that sample-based estimates can be analyzed using well-established statistical methods.

18.What is the significance of confidence intervals in statistical analysis?
Ans-The significance of confidence intervals (CIs) in statistical analysis lies in how they help quantify uncertainty and reliability in estimates of population parameters.
Significance of Confidence Intervals
1. Measures the reliability of estimates

A single point estimate (like a sample mean) doesn’t indicate how accurate it is.

A confidence interval shows the range of plausible values, helping us assess how reliable our estimate is.

Example: A 95% CI of (48, 52) is more precise (more reliable) than (40, 60).

2. Quantifies uncertainty

Sampling always involves variability.

Confidence intervals capture this uncertainty by showing how much the estimate might fluctuate due to random sampling.
3. Helps in decision-making

CIs guide whether to accept or reject certain claims about a population.

Example: If the 95% CI for mean income is (48k, 52k), and a company claims it’s 55k, the claim seems inconsistent with the data.

4. Supports hypothesis testing

Confidence intervals provide the same information as hypothesis tests, but in a more intuitive way.

If a 95% CI for a mean difference does not include 0, it suggests a statistically significant difference at the 5% level.

5. Compares groups effectively

CIs are useful in comparing means or proportions between groups.

If two CIs for group means don’t overlap much, it suggests a meaningful difference.

6. Communicates results clearly

CIs give readers a visual and interpretable measure of precision—more informative than just reporting a mean or p-value.

```python
2.Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)?

Python Code: Generate and Plot a Discrete Uniform Distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define parameters for the discrete uniform distribution
low, high = 1, 6  # outcomes from 1 to 6 (like a fair die)

# Generate random samples
samples = randint.rvs(low, high + 1, size=10000)

# Define possible outcomes
x = np.arange(low, high + 1)

# Compute PMF values
pmf = randint.pmf(x, low, high + 1)

# Plot the PMF
plt.stem(x, pmf, use_line_collection=True)
plt.title("Discrete Uniform Distribution (PMF)")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```

```python
3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?
Python Function Implementation
def bernoulli_pmf(x, p):
    """
    Calculate the probability mass function (PMF)
    of a Bernoulli distribution.

    Parameters:
    x (int): Outcome (0 or 1)
    p (float): Probability of success (0 <= p <= 1)

    Returns:
    float: Probability P(X = x)
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1.")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1.")

    return (p ** x) * ((1 - p) ** (1 - x))


# Example usage
p = 0.6
print("P(X=0):", bernoulli_pmf(0, p))
print("P(X=1):", bernoulli_pmf(1, p))
```

```python
5.Create a Poisson distribution and visualize it using Python?

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters
λ = 4  # average rate (mean number of events)
size = 10000  # number of samples

# Step 1: Generate random samples from a Poisson distribution
data = np.random.poisson(lam=λ, size=size)

# Step 2: Define range of possible values
x = np.arange(0, np.max(data) + 1)

# Step 3: Compute theoretical PMF values
pmf = poisson.pmf(x, λ)

# Step 4: Plot histogram of simulated data
plt.hist(data, bins=np.arange(-0.5, np.max(data) + 1.5, 1), density=True,
         alpha=0.6, color='skyblue', edgecolor='black', label='Simulated Data')

# Step 5: Plot the theoretical PMF
plt.plot(x, pmf, 'o-', color='red', label='Theoretical PMF')

# Step 6: Add titles and labels
plt.title(f"Poisson Distribution (λ = {λ})")
plt.xlabel("Number of Events (k)")
plt.ylabel("Probability")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
9.Write a Python function to calculate Z-scores from a dataset and plot them?

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def plot_z_scores(data):
    """
    Calculates Z-scores from a dataset and plots them.
    
    Parameters:
        data (list or array): Input dataset (numerical values)
    """
    # Step 1: Convert to NumPy array
    data = np.array(data)
    
    # Step 2: Calculate mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)
    
    # Step 3: Calculate Z-scores
    z_scores = (data - mean) / std_dev
    
    # Step 4: Print summary
    print(f"Mean: {mean:.2f}, Standard Deviation: {std_dev:.2f}")
    print("Z-scores:", np.round(z_scores, 2))
    
    # Step 5: Plot original data vs Z-scores
    plt.figure(figsize=(10, 5))

    # Original data distribution
    plt.subplot(1, 2, 1)
    sns.histplot(data, kde=True, color='skyblue')
    plt.title("Original Data Distribution")
    plt.xlabel("Data Values")
    plt.ylabel("Frequency")

    # Z-score distribution
    plt.subplot(1, 2, 2)
    sns.histplot(z_scores, kde=True, color='salmon')
    plt.title("Z-score Distribution")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")

    plt.tight_layout()
    plt.show()

    return z_scores

# Example dataset
data = [55, 60, 65, 70, 75, 80, 85, 90, 100, 110]

# Call the function
z = plot_z_scores(data)
```

```python
16.Create a confidence interval for a dataset using Python and interpret the result?

import numpy as np
from scipy import stats

# Step 1: Create a sample dataset
data = [52, 49, 55, 47, 51, 50, 53, 48, 54, 52]

# Step 2: Define confidence level
confidence = 0.95

# Step 3: Calculate statistics
n = len(data)
mean = np.mean(data)
std_dev = np.std(data, ddof=1)  # sample standard deviation
se = std_dev / np.sqrt(n)       # standard error

# Step 4: Get t-critical value for two-tailed 95% confidence interval
t_crit = stats.t.ppf((1 + confidence) / 2, df=n - 1)

# Step 5: Compute margin of error and confidence interval
margin_of_error = t_crit * se
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

# Step 6: Display results
print(f"Sample Mean = {mean:.2f}")
print(f"Sample Size = {n}")
print(f"Standard Deviation = {std_dev:.2f}")
print(f"Standard Error = {se:.3f}")
print(f"{int(confidence*100)}% Confidence Interval = ({lower_bound:.2f}, {upper_bound:.2f})")

# Step 7: Interpret the result
print(f"\nInterpretation: We are {int(confidence*100)}% confident that the true population mean lies between {lower_bound:.2f} and {upper_bound:.2f}.")
```

```python
22.Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution?
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate data from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=1000)  # mean=100, std=15

# Step 2: Calculate Range
data_range = np.max(data) - np.min(data)

# Step 3: Calculate IQR (Interquartile Range)
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Step 4: Print results
print(f"Minimum Value: {np.min(data):.2f}")
print(f"Maximum Value: {np.max(data):.2f}")
print(f"Range: {data_range:.2f}")
print(f"Q1 (25th percentile): {Q1:.2f}")
print(f"Q3 (75th percentile): {Q3:.2f}")
print(f"Interquartile Range (IQR): {IQR:.2f}")

# Step 5: Visualize with a boxplot
plt.figure(figsize=(6,5))
plt.boxplot(data, vert=True, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title("Boxplot of Normally Distributed Data")
plt.ylabel("Value")
plt.grid(alpha=0.4)
plt.show()
```

```python
23.Implement Z-score normalization on a dataset and visualize its transformation?

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Step 1: Generate a random dataset (normal distribution)
np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=1000)  # mean=100, std=15

# Step 2: Apply Z-score normalization
z_scores = zscore(data)

# Step 3: Verify transformation
print(f"Original Mean: {np.mean(data):.2f}, Standard Deviation: {np.std(data):.2f}")
print(f"After Z-Score Normalization → Mean: {np.mean(z_scores):.2f}, Standard Deviation: {np.std(z_scores):.2f}")

# Step 4: Visualization
plt.figure(figsize=(10,5))

# Original data histogram
plt.subplot(1,2,1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Original Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(alpha=0.4)

# Normalized data histogram
plt.subplot(1,2,2)
plt.hist(z_scores, bins=30, color='lightgreen', edgecolor='black', alpha=0.7)
plt.title("After Z-Score Normalization")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.grid(alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
24.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
distribution?

import numpy as np
from scipy.stats import skew, kurtosis
import matplotlib.pyplot as plt

# Function to calculate skewness and kurtosis
def analyze_distribution(mean=0, std=1, size=1000):
    # Step 1: Generate data from a normal distribution
    np.random.seed(42)
    data = np.random.normal(loc=mean, scale=std, size=size)
    
    # Step 2: Calculate skewness and kurtosis
    skewness = skew(data)
    kurt = kurtosis(data)  # By default, Fisher’s definition (normal dist → kurtosis = 0)
    
    # Step 3: Print results
    print(f"Mean: {np.mean(data):.2f}, Standard Deviation: {np.std(data):.2f}")
    print(f"Skewness: {skewness:.4f}")
    print(f"Kurtosis: {kurt:.4f}")
    
    # Step 4: Visualize the distribution
    plt.figure(figsize=(7,5))
    plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7, density=True)
    plt.title("Histogram of Normally Distributed Data")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.grid(alpha=0.4)
    plt.show()
    
    # Step 5: Interpretation
    if abs(skewness) < 0.1:
        print(" The distribution is approximately symmetric.")
    elif skewness > 0:
        print(" The distribution is positively skewed (right-tailed).")
    else:
        print("The distribution is negatively skewed (left-tailed).")
    
    if abs(kurt) < 0.1:
        print("The distribution has normal (mesokurtic) peakedness.")
    elif kurt > 0:
        print("The distribution is leptokurtic (sharper peak).")
    else:
        print("The distribution is platykurtic (flatter peak).")

# Call the function
analyze_distribution(mean=100, std=15, size=1000)
```
