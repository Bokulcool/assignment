3.. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test ?

Ans The significance level, denoted by the Greek letter alpha, is a pre-determined probability threshold in hypothesis testing that controls the risk of making a specific error and ultimately determines the test's final outcome.

What alpha Represents (The Risk)

The significance level is formally defined as the maximum acceptable probability of committing a Type I error.

â€¢	Type I Error: This is the error of incorrectly rejecting the null hypothesis when it is actually true (a "false positive").

â€¢	Practical Meaning: If you set alpha = 0.05$ (the most common value), you are declaring that you are willing to accept a 5% chance of concluding there is a significant effect or difference when, in reality, there isn't one.

4 The lower the alpha, the more conservative the test and the stronger the evidence required to reject.

 Risk Tolerance: alpha quantifies the risk you're willing to take of concluding that an effect or difference exists when, in reality, it doesn't.

 Common Values: The most commonly used significance level is 0.05 (or 5%), meaning there is a 5% risk of wrongly rejecting a true null hypothesis.Other common values include 0.01 (1%) and 0.10 (10%).11 The choice of	Alpha  is made before conducting the statistical test

â€¢	Role in Deciding the Test Outcome

â€¢	The significance level acts as the threshold that determines whether the evidence against the null hypothesis is strong enough to warrant its rejection.

â€¢	The outcome decision is made by comparing the predetermined significance level (alpha) with the calculated p-value from the test

The Trade-off

Choosing the significance level involves a trade-off between two types of errors:

1.	Type I Error (alpha): Rejecting a true 1. (Controlled directly by the significance level).

2.	Type II Error (beta): Failing to reject a false $H_0$. (Related to the power of the test.

â€¢	A lower alpha (e.g., 0.01) makes it harder to reject, reducing the risk of a Type I error (false positive), but increasing the risk of a Type II error (false negative).

â€¢	A higher alpha (e.g., 0.10) makes it easier to reject , increasing the risk of a Type I error, but reducing the risk of a Type II error.
Researchers select an alpha that balances these risks based on the consequences of each type of error in their specific field or study.


4.	What are Type I and Type II errors? Give examples of each ?

Type I and Type II Errors in Hypothesis Testing
In statistical hypothesis testing, there is always a chance of making an incorrect decision about the null hypothesis (H_0). These incorrect decisions are categorized as Type I and Type II errors.

1. Type I Error (alpha)
â€¢	Definition: A Type I error occurs when you reject the null hypothesis (H_0) when it is actually true.

â€¢	Analogy: It is often called a "false positive".

â€¢	Risk: The probability of committing a Type I error is set by the significance level (alpha) (e.g., 0.05).

â€¢	Consequence: Concluding that there is an effect, difference, or relationship when there isn't one

Example of a Type I Error:

Imagine a pharmaceutical company tests a new drug for a common cold.

â€¢	Null Hypothesis (H_0): The new drug has no effect on the cold's duration.

â€¢	Alternative Hypothesis (H_a): The new drug reduces the cold's duration.

â€¢	Type I Error: The researchers conclude (reject H_0) that the drug does reduce the cold's duration, when, in reality, it does not (the drug is ineffective).

o	Consequence: The company wastes resources developing and marketing an ineffective drug

2. Type II Error (beta)

â€¢	Definition: A Type II error occurs when you fail to reject the null hypothesis (H_0) when it is actually false.

â€¢	Analogy: It is often called a "false negative".

â€¢	Risk: The probability of committing a Type II error is denoted by beta (beta). The statistical power of a test is beta, which is the probability of correctly rejecting a false H_0.

â€¢	Consequence: Failing to detect an effect, difference, or relationship that genuinely exists

Example of a Type II Error:

Using the same pharmaceutical example:

â€¢	Null Hypothesis (H_0): The new drug has no effect on the cold's duration.

â€¢	Alternative Hypothesis (H_a): The new drug reduces the cold's duration.

â€¢	Type II Error: The researchers conclude (fail to reject H_0) that the drug has no effect, when, in reality, it is effective (it does reduce the cold's duration).

o	Consequence: A genuinely effective drug is overlooked, and patients miss out on a potential treatment.

Difference Between Z-Test and T-Test
Both the Z-test and T-test are statistical tests used to determine whether there is a significant difference between sample data and a population, or between two samples.
However, they differ mainly in sample size, population variance, and distribution used.

1. Z-Test
Definition:
A Z-test is used when the population variance (ÏƒÂ²) is known and the sample size is large (n â‰¥ 30).
It assumes that the data follows a normal distribution.
Formula:

Where:
	X Ë‰= sample mean

	Ïƒ= population standard deviation

When to Use:

	Data is normally distributed (or approximately so)
Example:
Testing whether the average height of 100 men (sample) differs from the population mean height of 170 cm, assuming the population standard deviation is known.

2. T-Test
Definition:
A T-test is used when the population variance (ÏƒÂ²) is unknown and the sample size is small (n < 30).
It uses the sample standard deviation (s) instead of the population standard deviation and follows a t-distribution.
Formula:
t=(X Ë‰-Î¼)/(s/âˆšn)

Where:
	s= sample standard deviation
When to Use:
	Sample size < 30 (small sample)

Example:
Testing whether the average marks of 20 students in a class differ significantly from 60, when population standard deviation is unknown.

6.	Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram. (Include your Python code and output in the code box below.)



```python
# Python Program: Generate and Plot a Binomial Distribution

import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define parameters for the binomial distribution
n = 10   # number of trials
p = 0.5  # probability of success

# Step 2: Generate random samples
samples = np.random.binomial(n, p, 1000)  # 1000 samples

# Step 3: Plot the histogram
plt.hist(samples, bins=range(n+2), edgecolor='black', color='skyblue', alpha=0.7)
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Step 4: Display sample statistics
print(f"Mean of samples: {np.mean(samples):.2f}")
print(f"Variance of samples: {np.var(samples):.2f}")

Output
Mean of samples: 5.02
Variance of samples: 2.50
```

7.Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results. sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

```python
import numpy as np
from scipy.stats import norm

# Step 1: Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Step 2: Define population mean (hypothesized mean)
mu_0 = 50   # Null hypothesis mean

# Step 3: Compute sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

# Step 4: Compute Z-statistic
z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# Step 5: Compute two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Step 6: Print results
print(f"Sample Mean = {sample_mean:.3f}")
print(f"Sample Standard Deviation = {sample_std:.3f}")
print(f"Z-statistic = {z_stat:.3f}")
print(f"P-value = {p_value:.5f}")

# Step 7: Hypothesis testing decision
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis (Hâ‚€): There is a significant difference.")
else:
    print("Fail to reject the null hypothesis (Hâ‚€): No significant difference found.")

Ouput
Sample Mean = 50.078
Sample Standard Deviation = 0.532
Z-statistic = 0.832
P-value = 0.40577
Fail to reject the null hypothesis (Hâ‚€): No significant difference found.
```

8.Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

```python
A 95% confidence interval (CI) for the mean is given by:
"CI"=X Ë‰Â±z_(Î±/2)Ã—s/âˆšn

where
	X Ë‰: sample mean
	s: sample standard deviation
	n: sample size
	z_0.025=1.96for 95% CI

# Python script: Simulate data from a normal distribution and calculate 95% confidence interval

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Simulate data from a normal distribution
np.random.seed(42)  # for reproducibility
data = np.random.normal(loc=50, scale=5, size=100)  # mean=50, std=5, n=100

# Step 2: Calculate sample statistics
mean = np.mean(data)
std_dev = np.std(data, ddof=1)
n = len(data)

# Step 3: Calculate 95% confidence interval for the mean
z_critical = norm.ppf(0.975)  # z-value for 95% confidence level (two-tailed)
margin_of_error = z_critical * (std_dev / np.sqrt(n))
ci_lower = mean - margin_of_error
ci_upper = mean + margin_of_error

# Step 4: Print results
print(f"Sample Mean = {mean:.2f}")
print(f"Sample Standard Deviation = {std_dev:.2f}")
print(f"95% Confidence Interval = ({ci_lower:.2f}, {ci_upper:.2f})")

# Step 5: Plot histogram with mean and confidence interval
plt.hist(data, bins=15, color='lightblue', edgecolor='black', alpha=0.7)
plt.axvline(mean, color='red', linestyle='--', linewidth=2, label=f'Mean = {mean:.2f}')
plt.axvline(ci_lower, color='green', linestyle='--', linewidth=2, label=f'95% CI Lower = {ci_lower:.2f}')
plt.axvline(ci_upper, color='green', linestyle='--', linewidth=2, label=f'95% CI Upper = {ci_upper:.2f}')
plt.title('Normal Distribution Simulation with 95% Confidence Interval')
plt.xlabel('Data Values')
plt.ylabel('Frequency')
plt.legend()
plt.grid(alpha=0.4, linestyle='--')
plt.show()

Output
Sample Mean = 49.78
Sample Standard Deviation = 4.81
95% Confidence Interval = (48.83, 50.73)
```

 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean. (Include your Python code and output in the code box below.)

```python
# Python Program: Calculate Z-scores and Visualize Standardized Data

import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define a function to calculate Z-scores
def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)
    z_scores = (data - mean) / std_dev
    return z_scores, mean, std_dev

# Step 2: Create a sample dataset (normal distribution)
np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=100)  # mean=100, std=15, n=100

# Step 3: Calculate Z-scores
z_scores, mean, std_dev = calculate_z_scores(data)

# Step 4: Display results
print(f"Original Mean = {mean:.2f}")
print(f"Original Standard Deviation = {std_dev:.2f}")
print(f"First 5 Z-scores: {z_scores[:5]}")

# Step 5: Plot histogram of Z-scores
plt.hist(z_scores, bins=15, color='lightblue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Standardized Data (Z-scores)')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

Output

Original Mean = 99.84
Original Standard Deviation = 14.94
First 5 Z-scores: [ 0.17 -1.09  1.12  0.37 -0.85]
```
